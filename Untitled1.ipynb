```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn import metrics 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
```

```python
data=pd.read_csv("breast-cancer.csv")
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop('id', axis=1, inplace=True)
data['diagnosis'] = data['diagnosis'].map({'B': 0, 'M': 1})
```

```python
data.head(500)
```

```python
X=data.drop("diagnosis",axis=1)
y=data.loc[:,"diagnosis"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.45, random_state=44, shuffle =True)
```

```python
lg=LogisticRegression(max_iter=500)
lg.fit(X_train,y_train)
lg.score(X_train,y_train)
```

```python
y_pred=lg.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
CM = confusion_matrix(y_test, y_pred)
print('Confusion Matrix is : \n', CM)

# drawing confusion matrix
sns.heatmap(CM, center = True)
plt.show()
```

```python
#Calculating Accuracy Score  : ((TP + TN) / float(TP + TN + FP + FN))
AccScore = accuracy_score(y_test, y_pred, normalize=False)
print('Accuracy Score is : ', AccScore)

#----------------------------------------------------
#(2*precision*Recall/precision+Recall)
F1Score = f1_score(y_test, y_pred, average='micro') #it can be : binary,macro,weighted,samples
print('F1 Score is : ', F1Score)

#----------------------------------------------------
#Calculating Precision Score : (Specificity) #(TP / float(TP + FN))  

RecallScore = recall_score(y_test, y_pred, average='micro')
print('Recall Score is : ', RecallScore)

#Calculating Precision Score : (Specificity) #(TP / float(TP + FP))  
precision_score=precision_score(y_test, y_pred, labels=None, pos_label=1, average='binary',sample_weight=None)
print("precision_score",precision_score)
```
